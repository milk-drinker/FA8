```python
import pandas as pd
import numpy as np

cv = lambda x: np.std(x, ddof = 1) / np.mean(x) * 100

df = pd.read_csv("https://raw.githubusercontent.com/Apress/data-analysis-and-visualization-using-python/master/Ch07/Salaries.csv")
df = df.fillna(df.mean())
```

```python
df_rank = df.groupby(by='rank')
perc = [.10, .20, .25, .30, .40, .50, .60, .70, .75, .80, .90]
```

```python
descStats = df_rank['salary'].describe(percentiles = perc, include ='all')
```

```python
descStats['Median'] = df_rank['salary'].apply(pd.Series.median)
descStats['Std. Error of mean'] = df_rank['salary'].apply(pd.Series.sem)
descStats['Coefficient of variation'] = df_rank['salary'].apply(cv)
descStats['Skewness'] = df_rank['salary'].apply(pd.Series.skew)
descStats['Kurtosis'] = df_rank['salary'].apply(pd.Series.kurtosis)
```
